```python
class ListNode(object):
    def __init__(self,x=0):
        self.val = x
        self.next = None
```

```python
class LinkedList(object):
    def __init__(self, *args):
        
        self.root = None
        self.length = 0
        if len(args)==1:
            self.append(args[0])
            
        
    def append(self,x):
        def element_append(lk,x):
            if lk.root == None:
                lk.root = ListNode(x)
            else:
                pcurr = lk.root
                pnext = lk.root.next
                while pnext != None:
                    pcurr = pnext
                    pnext = pnext.next
                pcurr.next = ListNode(x)
                
            lk.length += 1
                
        if type(x) == list:
            for i in range(len(x)):
                element_append(self,x[i])
        else:
            element_append(self,x)     
            
    def get_list(self):
        p = self.root
        arr = []
        while p != None:
            arr.append(p.val)
            p = p.next
        return arr
    
    def insert(self,pos,x): 
        # insert the element x before pos
        
        assert self.root != None
        assert pos >= 0 and pos < self.length
        
        node = ListNode(x)
        
        if pos == 0:
            node.next = self.root
            self.root = node
        else:
            prev = None
            curr = self.root
            for i in range(pos):
                prev = curr
                curr = curr.next
            node.next = curr
            prev.next = node
            
        self.length += 1
        
    def delete(self, pos):
        # delete the element at position pos
        
        assert pos >= 0 and pos < self.length
        assert self.root != None
        
        if pos == 0:
            self.root = self.root.next
        else:
            prev = None
            curr = self.root
            for i in range(pos):
                prev = curr
                curr = curr.next
            prev.next = curr.next
            
        self.length -= 1
            
                    
```
