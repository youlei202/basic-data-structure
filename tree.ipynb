```python
class TreeNode():
    def __init__(self,x=0):
        self.val = x
        self.left = None
        self.right = None
        
    def inorder(self,arr):
        if self == None:
            return
        if self.left != None:
            self.left.inorder(arr)
        arr.append(self.val)
        if self.right != None:
            self.right.inorder(arr)
            
    def preorder(self,arr):
        if self == None:
            return
        arr.append(self.val)
        if self.left != None:
            self.left.preorder(arr)
        if self.right != None:
            self.right.preorder(arr)
            
    def postorder(self,arr):
        if self == None:
            return
        if self.left != None:
            self.left.postorder(arr)
        if self.right != None:
            self.right.postorder(arr)
        arr.append(self.val)
        
    def find_successor(self):
        su_parent = self
        successor = su_parent.right
        
        while successor.left != None:
            su_parent = successor
            successor = successor.left
            
        return su_parent, successor
        


class BSTree():
    def __init__(self):
        self.root = None

        
    def insert(self,x):
        if self.root == None:
            self.root = TreeNode(x)
        else:
            curr_node = self.root
            parent = None 
            while curr_node != None:
                parent = curr_node
                if x <= curr_node.val:
                    curr_node = curr_node.left
                else:
                    curr_node = curr_node.right
            
            if x <= parent.val:
                parent.left = TreeNode(x)
            else:
                parent.right = TreeNode(x)
                
    def delete(self,x):
        if self.root == None:
            return
        else:
            curr_node = self.root
            parent = None
            while curr_node != None:
                if x < curr_node.val:
                    parent = curr_node # I used to put this line outside "if" wrongly
                    curr_node = curr_node.left
                elif x > curr_node.val:
                    parent = curr_node # the same as the last comment
                    curr_node = curr_node.right
                else:
                    break
            else:
                return
            
            if curr_node.left == None:
                if parent == None:
                    self.root = curr_node.right
                elif curr_node == parent.left: # mistake: I forgot to check whether curr_node belong to left or right of its parent
                    parent.left = curr_node.right
                else:
                    parent.right = curr_node.right
            elif curr_node.right == None:
                if parent == None:
                    self.root = curr_node.left
                elif curr_node == parent.left: # the same mistake as above
                    parent.left = curr_node.left
                else:
                    parent.right = curr_node.left # mistake: I used to write "parent.right" as "parent.left"
            else:
                su_parent, successor = curr_node.find_successor()
                su_parent.left = successor.right
                successor.right = su_parent
                successor.left = curr_node.left
                if parent == None:
                    self.root = successor
                elif curr_node == parent.left:
                    parent.left = successor
                else:
                    parent.right = successor
                
                
    def get_list(self, order = 'in'):
        if self.root == None:
            return []
        t = []
        if order == 'pre':
            self.root.preorder(t)
        elif order == 'post':
            self.root.postorder(t)
        else:
            self.root.inorder(t)
        return t
    
```
